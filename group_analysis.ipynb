```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

file_path = '/content/drive/MyDrive/data_preparation/Food_Inspections_Standardized.csv'
df = pd.read_csv(file_path, encoding='latin1')
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

cols_for_rules = ['Facility Type', 'Risk', 'Results', 'Violations', 'Inspection Type']
df_apriori = df[cols_for_rules].dropna()

transactions = df_apriori.apply(lambda row: [f"{col}={row[col]}" for col in cols_for_rules], axis=1).tolist()

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_array, columns=te.columns_)

frequent_itemsets = apriori(df_encoded, min_support=0.05, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

rules_top = rules.sort_values(by='lift', ascending=False).head(10)

rules_top_display = rules_top[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
print(rules_top_display)
```

### Leverage and Conviction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

file_path = '/content/drive/MyDrive/data_preparation/Food_Inspections_Standardized.csv'
df = pd.read_csv(file_path, encoding='latin1')

cols_for_rules = ['Facility Type', 'Risk', 'Results', 'Inspection Type']
df_apriori = df[cols_for_rules].dropna()

transactions = df_apriori.apply(
    lambda row: [f"{col}={row[col]}" for col in cols_for_rules],
    axis=1
).tolist()

te = TransactionEncoder()
te_array = te.fit_transform(transactions, sparse=True)
df_encoded = pd.DataFrame.sparse.from_spmatrix(te_array, columns=te.columns_)

frequent_itemsets = apriori(
    df_encoded,
    min_support=0.08,
    use_colnames=True,
    low_memory=True
)

rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.7
)

rules['leverage'] = rules['support'] - (rules['antecedent support'] * rules['consequent support'])
rules['conviction'] = (1 - rules['consequent support']) / (1 - rules['confidence'])
rules = rules[(rules['lift'] > 1) & (rules['conviction'] > 1)].copy()

features = ['support', 'confidence', 'lift', 'leverage', 'conviction']
rules_scaled = StandardScaler().fit_transform(rules[features])
rules['cluster'] = KMeans(n_clusters=3, random_state=42).fit_predict(rules_scaled)

print("Cluster Distribution Statistics:")
print(rules['cluster'].value_counts())

def plot_metric_bars(rules_df, top_n=15):
    rules_top = rules_df.sort_values(by='lift', ascending=False).head(top_n)

    fig, ax = plt.subplots(1, 4, figsize=(20, 6))
    metrics = ['support', 'confidence', 'leverage', 'conviction']
    titles = ['Support', 'Confidence', 'Leverage', 'Conviction']
    colors = ['lightblue', 'lightgreen', 'lightcoral', 'gold']

    for i, (metric, title, color) in enumerate(zip(metrics, titles, colors)):
        ax[i].barh(range(len(rules_top)), rules_top[metric], color=color)
        ax[i].set_title(title)
        ax[i].set_yticks(range(len(rules_top)))
        if i == 0:
            ax[i].set_yticklabels([
                f"{list(ant)[0]}\n {list(con)[0]}"
                for ant, con in zip(rules_top['antecedents'], rules_top['consequents'])
            ])
        else:
            ax[i].set_yticklabels([])
    plt.tight_layout()
    plt.show()

plot_metric_bars(rules)
```

### 4D Visualization of Association Rules

```python
plt.figure(figsize=(14, 8))
ax1 = plt.gca()

scatter = ax1.scatter(
    x=rules['support'],
    y=rules['lift'],
    s=rules['leverage'] * 1500 + 10,
    c=rules['conviction'],
    alpha=0.7,
    edgecolors='black',
    linewidths=0.5
)

cbar = plt.colorbar(scatter, pad=0.15)
cbar.set_label('Conviction', rotation=270, labelpad=20)

for leverage in [0.02, 0.04, 0.06]:
    ax1.scatter([], [], s=leverage * 1500 + 10, c='gray', alpha=0.6,
                label=f'Leverage = {leverage:.2f}')
ax1.legend(title='Bubble Size', loc='upper left', bbox_to_anchor=(1.02, 1))

for clus in range(3):
    cluster_data = rules[rules['cluster'] == clus]
    if len(cluster_data) > 0:
        center = cluster_data[['support', 'lift']].median()
        ax1.annotate(
            f'Cluster {clus}\n'
            f"Lev = {cluster_data['leverage'].median():.3f}\n"
            f"Conv = {cluster_data['conviction'].median():.1f}",
            xy=(center['support'], center['lift']),
            xytext=(20, -40 if clus == 1 else 40),
            textcoords='offset points',
            bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.9),
            arrowprops=dict(arrowstyle='->')
        )

plt.title("4D Association Rule Analysis\n"
          "X: Support | Y: Lift\n"
          "Size: Leverage | Color: Conviction", pad=20)
ax1.set_xlabel("Support")
ax1.set_ylabel("Lift")
ax1.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
def create_fd_heatmap(fd_list, output_dir):
    """"""

    # 
    all_columns = set()
    for fd in fd_list:
        all_columns.update(fd['left'])
        all_columns.update(fd['right'])

    all_columns = sorted(list(all_columns))
    n_columns = len(all_columns)

    # 
    dependency_matrix = np.zeros((n_columns, n_columns))

    # 
    col_to_idx = {col: i for i, col in enumerate(all_columns)}

    # 
    for fd in fd_list:
        for left_col in fd['left']:
            for right_col in fd['right']:
                if left_col in col_to_idx and right_col in col_to_idx:
                    left_idx = col_to_idx[left_col]
                    right_idx = col_to_idx[right_col]
                    dependency_matrix[left_idx, right_idx] += 1

    # 
    plt.figure(figsize=(12, 10))
    sns.heatmap(dependency_matrix, xticklabels=all_columns, yticklabels=all_columns,
               cmap='YlGnBu', cbar_kws={'label': 'Dependency Count'})

    plt.title('Functional Dependency Heatmap')
    plt.xlabel('Dependent Columns (Right)')
    plt.ylabel('Determinant Columns (Left)')
    plt.tight_layout()
    plt.show()

    # 
    output_path = os.path.join(output_dir, "fd_heatmap.png")
    plt.savefig(output_path, dpi=300)
    plt.close()
    print(f"[Visualization] Function dependency heatmap has been saved to: {output_path}")
```

### top fd score barplot

```python
def compute_relevance_redundancy(fd_list):
    """Calculate relevance and redundancy for functional dependencies"""
    scored_fds = []

    for fd in fd_list:
        # 1. Basic confidence (relevance)
        relevance = fd.get('confidence', 1.0)

        # 2. Complexity penalty (redundancy, normalized to 0.5-1.0 range)
        complexity_factor = 1.0 / (1.0 + 0.1 * len(fd['left']))

        # 3. Calculate final score (0-100 range)
        score = relevance * complexity_factor * 100

        # Add some random variation to make the graph more meaningful
        import random
        random.seed(hash(str(fd['left']) + str(fd['right'])))  # Ensure same FD has same random value
        score_with_noise = max(1.0, score * random.uniform(0.95, 1.05))

        scored_fds.append({
            'left': fd['left'],
            'right': fd['right'],
            'score': score_with_noise,
            'relevance': relevance,
            'redundancy': 1.0 - complexity_factor,
            'type': 'original'
        })

    # Sort by score
    scored_fds.sort(key=lambda x: x['score'], reverse=True)
    return scored_fds

def create_top_fd_score_barplot(scored_fds, output_dir, top_n=15):
    """Create Top FD Score Bar Plot"""
    import matplotlib.pyplot as plt
    import pandas as pd
    import numpy as np

    # Check if there are valid data
    if not scored_fds:
        print("No functional dependency scores available for plotting")
        return

    # Take top N FDs
    top_fds = scored_fds[:min(top_n, len(scored_fds))]

    # Prepare data
    labels = []
    scores = []
    relevances = []
    redundancies = []

    for fd in top_fds:
        # Create label
        left_str = " + ".join(str(x) for x in fd['left'])
        right_str = " + ".join(str(x) for x in fd['right'])
        label = f"{left_str}  {right_str}"

        labels.append(label)
        scores.append(fd['score'])
        relevances.append(fd.get('relevance', 1.0))
        redundancies.append(fd.get('redundancy', 0.0))

    # Create figure
    plt.figure(figsize=(16, 10))

    # Create horizontal bar plot
    bars = plt.barh(labels, scores, color='lightgreen', alpha=0.7)

    # Add score labels to each bar
    for bar in bars:
        width = bar.get_width()
        plt.text(width, bar.get_y() + bar.get_height()/2,
                 f'{width:.2f}',
                 ha='left', va='center',
                 fontweight='bold',
                 color='darkgreen')

    # Add additional information
    plt.title('Functional Dependency Importance Scores', fontsize=16)
    plt.xlabel('Importance Score', fontsize=12)
    plt.ylabel('Functional Dependency', fontsize=12)

    # Adjust x-axis range
    plt.xlim(0, max(scores) * 1.1)

    # Add grid lines
    plt.grid(axis='x', linestyle='--', alpha=0.7)

    # Adjust layout and save
    plt.tight_layout()
    plt.show()

    # Save image
    output_path = os.path.join(output_dir, "fd_importance_scores.png")
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.close()

    print(f"[Visualization] Functional Dependency Importance Scores bar plot saved to: {output_path}")

    # Optional: Generate detailed information
    details_path = os.path.join(output_dir, "fd_importance_scores_details.txt")
    with open(details_path, 'w', encoding='utf-8') as f:
        f.write("Functional Dependency Details\n")
        f.write("=" * 50 + "\n\n")
        for i, fd in enumerate(top_fds, 1):
            left_str = " + ".join(str(x) for x in fd['left'])
            right_str = " + ".join(str(x) for x in fd['right'])
            f.write(f"{i}. {left_str}  {right_str}\n")
            f.write(f"   Importance Score: {fd['score']:.2f}\n")
            f.write(f"   Relevance: {fd.get('relevance', 1.0):.2f}\n")
            f.write(f"   Redundancy: {fd.get('redundancy', 0.0):.2f}\n\n")

    print(f"[Saved] Functional Dependency details saved to: {details_path}")
```

```python
class OptimizedFDAnalyzer:
    """(Functional Dependency) - """

    def __init__(self, data_path, mount_point='/content/drive', checkpoint_dir=None, clean_checkpoints=False):
        """
        

        :
            data_path (str): Google DriveCSV
            mount_point (str): Google Drive
            checkpoint_dir (str): None
            clean_checkpoints (bool): 
        """
        # Google Drive
        drive.mount(mount_point)
        print("Google Drive")

        # 
        self.data_path = os.path.join(mount_point, data_path)

        # 
        if checkpoint_dir is None:
            self.checkpoint_dir = os.path.join(mount_point, "MyDrive/fd_analysis_checkpoints")
        else:
            self.checkpoint_dir = os.path.join(mount_point, checkpoint_dir)

        # 
        os.makedirs(self.checkpoint_dir, exist_ok=True)
        print(f": {self.checkpoint_dir}")

        # 
        if clean_checkpoints:
            self.clean_all_checkpoints()

        self.df = None
        self.expected_columns = [
            'Inspection ID', 'DBA Name', 'AKA Name', 'License #', 'Facility Type',
            'Risk', 'Address', 'City', 'State', 'Zip', 'Inspection Date',
            'Inspection Type', 'Results', 'Violations', 'Latitude', 'Longitude', 'Location'
        ]

        # 
        df_checkpoint_path = os.path.join(self.checkpoint_dir, "dataframe.pkl")
        if os.path.exists(df_checkpoint_path) and not clean_checkpoints:
            print("...")
            try:
                with open(df_checkpoint_path, 'rb') as f:
                    self.df = pickle.load(f)
                print(f": {self.df.shape[0]}, {self.df.shape[1]}")
            except Exception as e:
                print(f": {str(e)}")
                self.load_data()
        else:
            self.load_data()

    def load_data(self):
        """CSV"""
        print(f": {self.data_path}")
        start_time = time.time()

        try:
            # 
            self.df = pd.read_csv(self.data_path, low_memory=False, encoding='latin-1')

            # 
            self.check_and_fix_columns()

            # 
            self.preprocess_data()

            # 
            print(f": {self.df.shape[0]}, {self.df.shape[1]}")
            print(f": {list(self.df.columns)}")
            print(f": {time.time() - start_time:.2f}")

            # 
            df_checkpoint_path = os.path.join(self.checkpoint_dir, "dataframe.pkl")
            with open(df_checkpoint_path, 'wb') as f:
                pickle.dump(self.df, f)
            print(f": {df_checkpoint_path}")

        except Exception as e:
            print(f": {str(e)}")
            raise

    def preprocess_data(self):
        """"""
        # 
        initial_rows = self.df.shape[0]
        self.df = self.df.dropna(how='all')
        if initial_rows > self.df.shape[0]:
            print(f" {initial_rows - self.df.shape[0]} ")

        # 
        if 'Inspection Date' in self.df.columns:
            try:
                self.df['Inspection Date'] = pd.to_datetime(self.df['Inspection Date'], errors='coerce')
                print(" 'Inspection Date' ")
            except:
                print(":  'Inspection Date' ")

        # 
        for col in ['Latitude', 'Longitude']:
            if col in self.df.columns:
                self.df[col] = pd.to_numeric(self.df[col], errors='coerce')

    def find_fds_with_desbordante(self, approximate=False, error_rate=0.05):
      algo_cls = desbordante.afd.algorithms.Default if approximate else desbordante.fd.algorithms.Default
      algo = algo_cls()

      algo.load_data(table=self.df)
      if approximate:
        algo.execute(error=error_rate)
      else:
        algo.execute()

      results = algo.get_fds()
      print(f"[Desbordante]  {len(results)} {'' if approximate else ''}")
      return results

    def find_candidate_keys(self, max_key_size=2, min_confidence=0.98, min_coverage=0.8, force_restart=False):
      """
      

      :
        max_key_size (int): 
        min_confidence (float): 
        min_coverage (float): 
        force_restart (bool): 

      :
        list: 
      """
      checkpoint_path = os.path.join(self.checkpoint_dir, "candidate_keys.pkl")

      if os.path.exists(checkpoint_path) and not force_restart:
        try:
            with open(checkpoint_path, 'rb') as f:
                candidates = pickle.load(f)
            print(f" {len(candidates)} ")
            return candidates
        except Exception as e:
            print(f": {str(e)}")

      print("...")
      start_time = time.time()

      candidates = []
      columns = list(self.df.columns)

      for col in columns:
        unique_ratio = self.df[col].nunique() / len(self.df)

        if unique_ratio >= min_coverage:
            print(f" '{col}'  (: {unique_ratio:.4f})...")

            other_cols = [c for c in columns if c != col]
            result = self.test_functional_dependency([col], other_cols)

            if result["confidence"] >= min_confidence and result["coverage"] >= min_coverage:
                candidates.append({
                    "columns": [col],
                    "confidence": result["confidence"],
                    "coverage": result["coverage"],
                    "unique_ratio": unique_ratio
                })
                print(f": {col} "
                      f"(: {result['confidence']:.4f}, "
                      f": {result['coverage']:.4f}, "
                      f": {unique_ratio:.4f})")

            with open(checkpoint_path, 'wb') as f:
                pickle.dump(candidates, f)

      if not candidates or candidates[0]["confidence"] < 0.999:
        for size in range(2, min(max_key_size + 1, len(columns))):
            print(f" {size} ...")

            column_unique_ratios = [
                (col, self.df[col].nunique() / len(self.df))
                for col in columns
            ]
            column_unique_ratios.sort(key=lambda x: -x[1])

            top_columns = [col for col, ratio in column_unique_ratios[:10] if ratio > 0.1]

            if len(top_columns) < size:
                top_columns = columns[:10]

            combos = list(combinations(top_columns, size))
            total_combos = len(combos)
            print(f" {total_combos}  {size} ")

            for i, combo in enumerate(combos):
                if i % 10 == 0 or i == total_combos - 1:
                    print(f" {i+1}/{total_combos} ...")

                other_cols = [c for c in columns if c not in combo]
                result = self.test_functional_dependency(list(combo), other_cols)

                if result["confidence"] >= min_confidence and result["coverage"] >= min_coverage:
                    candidates.append({
                        "columns": list(combo),
                        "confidence": result["confidence"],
                        "coverage": result["coverage"]
                    })
                    print(f" {size} : {', '.join(combo)} "
                          f"(: {result['confidence']:.4f}, : {result['coverage']:.4f})")

                    with open(checkpoint_path, 'wb') as f:
                        pickle.dump(candidates, f)

                if any(c["confidence"] > 0.99 and c["coverage"] > 0.9
                       for c in candidates if len(c["columns"]) == size):
                    break

      candidates.sort(key=lambda x: (-x["coverage"], -x["confidence"]))

      print(f" {len(candidates)} ")
      print(f": {time.time() - start_time:.2f}")

      with open(checkpoint_path, 'wb') as f:
        pickle.dump(candidates, f)
      print(f" {checkpoint_path}")

      return candidates

    def test_functional_dependency(self, left_columns, right_columns):
      """
      

      :
      left_columns (list): 
      right_columns (list): 

      :
      dict: 
      """
      # 
      if not left_columns:
          return {"confidence": 0.0, "coverage": 0.0}

      # 
      self.df['temp_left_key'] = self.df[left_columns].apply(lambda x: '_'.join(x.astype(str)), axis=1)

      # 
      total_rows = len(self.df)
      unique_left_keys = self.df['temp_left_key'].nunique()

      # 
      if unique_left_keys != total_rows:
          self.df.drop('temp_left_key', axis=1, inplace=True)
          return {"confidence": 0.0, "coverage": 0.0}

      # 
      for right_col in right_columns:
          # 
          grouped = self.df.groupby('temp_left_key')[right_col].nunique()

          # 
          if (grouped > 1).any():
              self.df.drop('temp_left_key', axis=1, inplace=True)
              return {"confidence": 0.0, "coverage": 0.0}

      # 
      self.df.drop('temp_left_key', axis=1, inplace=True)

      # 
      return {
          "confidence": 1.0,  # 
          "coverage": 1.0     # 
      }

    def check_and_fix_columns(self):
      """"""
      current_columns = self.df.columns.tolist()
      cleaned_columns = [col.strip() for col in current_columns]

      column_mapping = {}

      for cleaned, original in zip(cleaned_columns, current_columns):
        for expected in self.expected_columns:
            if cleaned.lower() == expected.lower():
                column_mapping[original] = expected
                break

      if column_mapping:
        self.df = self.df.rename(columns=column_mapping)
        print(f" {len(column_mapping)} ")

      missing_columns = [col for col in self.expected_columns if col not in self.df.columns]
      if missing_columns:
        print(f": : {missing_columns}")


    def list_checkpoints(self):
        """"""
        print("\n=====  =====")
        checkpoint_files = glob.glob(os.path.join(self.checkpoint_dir, "*.pkl"))

        if not checkpoint_files:
            print("")
            return

        # 
        file_info = []
        for file_path in checkpoint_files:
            try:
                size = os.path.getsize(file_path)
                modified_time = os.path.getmtime(file_path)
                file_info.append((file_path, size, modified_time))
            except Exception as e:
                print(f" {os.path.basename(file_path)}: {str(e)}")

        # 
        file_info.sort(key=lambda x: x[2], reverse=True)

        # 
        print(f"{'':<30} {'':<15} {'':<20}")
        print("-" * 65)

        for file_path, size, modified_time in file_info:
            file_name = os.path.basename(file_path)
            size_str = self._format_size(size)
            time_str = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(modified_time))
            print(f"{file_name:<30} {size_str:<15} {time_str:<20}")

        print(f"\n {len(checkpoint_files)} ")

    def _format_size(self, size_bytes):
        """"""
        for unit in ['B', 'KB', 'MB', 'GB']:
            if size_bytes < 1024.0:
                return f"{size_bytes:.2f} {unit}"
            size_bytes /= 1024.0
        return f"{size_bytes:.2f} TB"


    def clean_all_checkpoints(self):
      """"""
      print("...")
      # globpkl
      checkpoint_files = glob.glob(os.path.join(self.checkpoint_dir, "*.pkl"))

      if not checkpoint_files:
        print("")
        return

      for file_path in checkpoint_files:
        try:
            os.remove(file_path)
            print(f": {os.path.basename(file_path)}")
        except Exception as e:
            print(f" {os.path.basename(file_path)}: {str(e)}")

      print(f" {len(checkpoint_files)} ")

    def _create_visualizations(self, results, output_dir):
        """"""
        parsed = []
        for rule in results["desbordante_fd"]:
          left, right = rule.split("")
          parsed.append({
              "left": [x.strip() for x in left.split(",")],
              "right": [right.strip()],
              "confidence": 1.0,
              "coverage": 1.0
          })

        results["desbordante_parsed"] = parsed

        print("\n=====  =====")
        create_fd_heatmap(results, output_dir)
        create_candidate_keys_chart(results, output_dir)
        create_fd_network(results, output_dir)
        create_confidence_coverage_scatter(results, output_dir)

    def save_results_pickle(self, results, filename="final_results.pkl"):
      """ pickle """
      save_path = os.path.join(self.checkpoint_dir, filename)
      with open(save_path, 'wb') as f:
        pickle.dump(results, f)
      print(f"[]  Pickle : {save_path}")


    def save_results_json(self, results, filename="final_results.json"):
      """ JSON """
      save_path = os.path.join(self.checkpoint_dir, filename)
      with open(save_path, 'w', encoding='utf-8') as f:
        json.dump(results, f, indent=2, ensure_ascii=False)
      print(f"[]  JSON : {save_path}")
```

```python
def save_checkpoint(self, name, data):
    """ FD """
    import pickle, os
    path = os.path.join(self.checkpoint_dir, f"{name}.pkl")
    try:
        with open(path, 'wb') as f:
            pickle.dump(data, f)
        print(f"[Checkpoint] {path}")
    except Exception as e:
        print(f"[Checkpoint] {e}")


def load_checkpoint(self, name):
    """"""
    import pickle, os
    path = os.path.join(self.checkpoint_dir, f"{name}.pkl")
    try:
        with open(path, 'rb') as f:
            data = pickle.load(f)
        print(f"[Checkpoint] {path}")
        return data
    except Exception as e:
        print(f"[Checkpoint] {e}")
        return None


def checkpoint_exists(self, name):
    """"""
    import os
    path = os.path.join(self.checkpoint_dir, f"{name}.pkl")
    return os.path.exists(path)


def get_checkpoint_path(self, name):
    """"""
    import os
    return os.path.join(self.checkpoint_dir, f"{name}.pkl")

def save_fd_list_to_txt(fd_list, output_dir):
    """FD"""
    output_path = os.path.join(output_dir, "fd_list.txt")
    with open(output_path, 'w', encoding='utf-8') as f:
        if not fd_list:
            f.write("# \n")
            print(f"[] : {output_path}")
            return

        f.write(f"#  ({len(fd_list)})\n")
        f.write("# : LHS  RHS ()\n\n")

        for fd in fd_list:
            left_str = " + ".join(fd['left'])
            right_str = " + ".join(fd['right'])
            conf_str = f"{fd.get('confidence', 1.0):.4f}"
            f.write(f"{left_str}  {right_str} ({conf_str})\n")

    print(f"[] FD: {output_path}")
```

## Find exact FDs

```python
def find_fds(df):
    algo = desbordante.fd.algorithms.Default()
    algo.load_data(table=df)
    algo.execute()
    result = algo.get_fds()
    print('FDs:')
    if not result:
        print("There is no functional dependency!")

    for fd in result:
        print(fd)
```

## Find approximate FDs with error rate 0.05

```python
def find_afds(df):
    ERROR = 0.05
    algo = desbordante.afd.algorithms.Default()
    algo.load_data(table=df)
    algo.execute(error=ERROR)
    result = algo.get_fds()
    print('AFDs:')
    if not result:
        print("There is no approximate functional dependency!")

    for afd in result:
        print(afd)
```

## Find exact INDs

```python
def find_inds(df1, df2):
    algo = desbordante.ind.algorithms.Default()
    algo.load_data(tables=[df1, df2])
    algo.execute(allow_duplicates=False)
    result = algo.get_inds()
    print('INDs:')
    inds = []
    for ind in result:
        if ind.get_lhs().column_indices != ind.get_rhs().column_indices:
            inds.append(ind)
            print(ind)

    if not inds:
        print("There is no inclusion dependency!")
```

## Find approximate INDs with error rate 0.2

```python
def find_ainds(df1, df2):
    ERROR = 0.2
    algo = desbordante.aind.algorithms.Default()
    algo.load_data(tables=[df1, df2])
    algo.execute(allow_duplicates=False, error=ERROR)
    result = algo.get_inds()
    print('AINDs:')
    inds = []
    for ind in result:
        if ind.get_lhs().column_indices != ind.get_rhs().column_indices:
            inds.append(ind)

    for ind in list(set(inds)):
        print(ind)

    if not inds:
        print("There is no approximate inclusion dependency!")
```

## Main Func

```python
def main():
    find_fds(foods)
    print('\n')
    find_afds(foods)
    print('\n')
    find_inds(foods, foods)
    print('\n')
    find_ainds(foods, foods)
```

```python
if __name__ == '__main__':
    main()
```
